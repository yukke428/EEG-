```python
%matplotlib inline
#FFT
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ptick
from matplotlib.ticker import ScalarFormatter
import os
from scipy import signal
from scipy.fftpack import fft,fftfreq

path="/Users/yusuke/Desktop/InputTime_NormalStanding/"
outpath="/Users/yusuke/Desktop/OutputFFT_NormalStanding/"
N=512
i=0
dt=3.025*pow(10,-10)
t=np.linspace(1,N,N)*dt-dt

han=signal.hann(N)

class FixedOrderFormatter(ScalarFormatter):
    def __init__(self, order_of_mag=0, useOffset=True, useMathText=True):
        self._order_of_mag = order_of_mag
        ScalarFormatter.__init__(self, useOffset=useOffset, 
                                 useMathText=useMathText)
    def _set_orderOfMagnitude(self, range):
        self.orderOfMagnitude = self._order_of_mag
        
    
def readCSV(name):
    Inpath=os.path.join(path,name)
    csv_file=open(Inpath,"r")
    f=csv.reader(csv_file,delimiter=",",doublequote=True,lineterminator="\n\n")
    
    Time=[]
    Input=[]
    Output=[]
    n=0
    
    for row in f:
        if n<N:
            Time.append(row[0])
            Input.append(row[1])
            Output.append(row[3])
            n+=1
    
    Time=np.array(Time).astype(np.float64)
    Input=np.array(Input).astype(np.float64)
    Output=np.array(Output).astype(np.float64)
    Input=[x*y for (x,y) in zip(Input,han)]
    Output=[x*y for (x,y) in zip(Output,han)]
    return Time,Input,Output
    
def writeCSV(name,Freq,Input_FFT_real,Input_FFT_img,Input_FFT_abs,Output_FFT_real,Output_FFT_img,Output_FFT_abs):
    Outpath=os.path.join(outpath,name)
    with open(Outpath,"w") as f:
        writer=csv.writer(f,lineterminator="\n")
        j=0
        for f in Freq:
            if j<len(Freq):
                write_list=[Freq[j],Input_FFT_real[j],Input_FFT_img[j],Input_FFT_abs[j],Output_FFT_real[j],Output_FFT_img[j],Output_FFT_abs[j],"\n"]
                writer.writerow(write_list)
      
                j+=1
    
def TimeToFreq(Input,Output):
    Input_FFT=np.fft.fft(Input)/(N/2)
    Output_FFT=np.fft.fft(Output)/(N/2)
    Input_FFT[0]=Input_FFT[0]/2
    Output_FFT[0]=Output_FFT[0]/2
    
    Freq=fftfreq(N,dt)
    return Freq,Input_FFT,Output_FFT

directory=os.listdir(path)
directory.pop(0)

if __name__=="__main__":

    for excel in directory:
        Time,Input,Output=readCSV("data%d.csv"% i)
        Freq,Input_FFT,Output_FFT=TimeToFreq(Input,Output)
        Out=np.fft.ifft(Output_FFT)
        writeCSV("data%d.csv"%i,Freq,Input_FFT.real,Input_FFT.imag,abs(Input_FFT),Output_FFT.real,Output_FFT.imag,abs(Output_FFT))
        i+=1
    print("success")
```
